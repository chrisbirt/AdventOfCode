```python
import re

X = 0
Y = 1

def init_cave(puzzle_input):

    cave = []
    rock_paths = []

    min_x = 0
    max_x = 0
    max_y = 0
    
    # reformat puzzle input
    # "498,4 -> 498,6 -> 496,6" becomes [[498,4],[498,6],[496,6]] 
    
    for path in puzzle_input:
        rock_paths.append([[int(point[X]), int(point[Y])] for point in [points.split(',') for points in re.split(' -> ', path)]])
        
    # determine the boundaries of cave (min/max x, max y)
    # assume it neets to be twice as wide as deep (since the sand will form as a pyramid)
    # hence, find the deepest point
   
    for path in rock_paths:
        max_y = max(max_y, max([point[Y] for point in path]))
        
    max_y = max_y + 1
    min_x = 500 - max_y - 1
    max_x = 500 + max_y + 1
    
    # now initialize the cave ('.')

    cave = [x[:] for x in [['.'] * (max_x-min_x)] * (max_y+2)]  
    
    # populate with rocks ('#')         
    
    for path in rock_paths:
        
        for i in range(len(path)-1):
            
            # from/to could be right to left, down to up, so re-order the coordinates
            from_x, to_x = sorted([path[i][X], path[i+1][X]])
            from_y, to_y = sorted([path[i][Y], path[i+1][Y]])

            # populate with rocks between from/to
            for x in range(from_x, to_x+1):
                for y in range(from_y, to_y+1):
                    cave[y][x-min_x] = '#'
    
    return cave, max_y


def flow_sand(cave, max_y):
    
    sand = [500,0]
    sand_count = 0
    result_1 = 0

    while True:

        # get the next three possible points

        next_sand = cave[sand[Y]+1][(sand[X]-1)-min_x:(sand[X]+2)-min_x]

        # Has sand come to rest?
        
        if (sand[Y] == max_y) or ('.' not in next_sand):

            # place the sand in the cave
            
            cave[sand[Y]][sand[X]-min_x] = 'O'
            sand_count += 1

            # is this the first grain to reach the boundary?
            
            if (sand[Y] == max_y) and (result_1 == 0):
                result_1 = sand_count-1 if result_1 == 0 else result_1
            
            # is this grain of sand at the top of the cave?
            
            if sand == [500,0]:
                cave[sand[Y]][sand[X]-min_x] = 'O'
                return(result_1, sand_count)

            # reset sand
            sand = [500,0] 

        # move the grain of sand
        
        elif next_sand[1] == '.':
            sand = [sand[X], sand[Y]+1]

        elif next_sand[0] == '.':
            sand = [sand[X]-1, sand[Y]+1]
        
        elif next_sand[2] == '.':
            sand = [sand[X]+1, sand[Y]+1]
            
 
with open('.\\input\\day14.txt', 'r') as f:
    # using splitlines removes the newline characters from the input
    puzzle_input = f.read().splitlines()
    
# build the cave
cave, max_y = init_cave(puzzle_input)

# let the sand flow!
result_1, result_2 = flow_sand(cave, max_y)

print('Day 14, part 1:', result_1)
print('Day 14, part 2:', result_2)
```
